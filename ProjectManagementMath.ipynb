#PV=FV/(1+r)<sup>n</sup>
PV= Present Value<br>
FV= Future Value<br>
r= discount/incriment rate<br>
n= time period<br>

```python
FV=int(input("Enter the future value, FV= "))
r=int(input("r= "))
n=int(input("Time Peroid, n= "))

print("The Prasent value of "+str(FV)+" is , PV= ", (FV/(1+(r/100))**n))
```

#Net Present Value(NPV)

∑(CFn / (1 + i)<sup>n</sup>) – Initial Investment

CFn= Future value<br>
i=rate
n= time period


```python

CF=int(input("Enter the future value , CV= "))
ii=int(input("Enter the initial value= "))
r=int(input("r= "))
n=int(input("Time Peroid, n= "))

sum=0
for i in range(1,n+1):
  sum=sum+(CF/(1+(r/100))**i)
  print("PV",i," =",(CF/(1+(r/100))**i))
print("The NET PRESENT VALUE is, NPV= ",(sum-ii))
```

#Benefit-Cost Ratio (BCR)

Benefit-Cost Ratio = PV of Expected Benefits / PV of Expected Costs

```python
inv=int(input("Investment: "))
print("Incremental benefits")
FV=list(map(int,input().split()))
r=int(input("r= "))
sum=0
for i in range(len(FV)):
  print("benefits in "+str(i+1)+" is , PV= ", (FV[i]/(1+(r/100))**(i+1)))
  sum=sum+(FV[i]/(1+(r/100))**(i+1))

print("Benefit-Cost Ratio =", (sum/inv))
```

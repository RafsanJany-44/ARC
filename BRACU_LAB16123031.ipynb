```python
n=int(input("Enter the number :"))

def show_palindromic_triangle(n):
    a=2*n
    for i in range(1,n+1):
        print(" "*a,end='')
        a-=2
        for j in range(1,i):
            print(str(j)+" ",end='')
        for k in range(i,0,-1):
            print(str(k)+" ",end='')
        print(end='\n')
show_palindromic_triangle(n)
```

Write a python function that takes a list as an argument. Your task is to create a new list where each element can be present at max 2 times. Inside the function, print the number of elements removed from the given list. Finally, return the new list and print the result.

=====================================================

Hint: You may use list_name.count(element) to count the total number of times an element is in a list. list_name is your new list for this problem.

=====================================================

Function Call:
function_name([1, 2, 3, 3, 3, 3, 4, 5, 8, 8])
Output:
Removed: 2
[1, 2, 3, 3, 4, 5, 8, 8]
================================
Function Call:
function_name([10, 10, 15, 15, 20])
Output:
Removed: 0
[10, 10, 15, 15, 20]

```python
def fun(k):
  l=list(set(k))
  final=[]
  hishab=0
  for i in l:
    if k.count(i)>=2:
      final.append(i)
      final.append(i)
      hishab=hishab+k.count(i)-2
    else:
      final.append(i)
  return (hishab,final)


res=fun([1, 2, 3, 3, 3, 3, 4, 5, 8, 8])
print("Removed: "+str(res[0]))
print(res[1])
```

Input:
"*"
5.5
2.5
Function Call:
function_name("*", 5.5, 2.5)

13.75

```python
def basic_calc(tup):
  if tup[0]=="+":
    return tup[1]+tup[2]
  elif tup[0]=="-":
    return tup[1]-tup[2]
  elif tup[0]=="*":
    return tup[1]*tup[2]
  elif tup[0]=="/":
    return tup[1]/tup[2]

oprt=input()
oprd1=float(input())
oprd2=float(input())
print(basic_calc((oprt, oprd1, oprd2)))
```

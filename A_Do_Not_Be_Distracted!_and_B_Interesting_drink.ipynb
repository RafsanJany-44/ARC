```python
 def raf(n,x): 
  m=[x[0]]
  for i in range(1,len(x)):
    if x[i]!=x[i-1]:
      if x[i] in m:
        break
      else:
        m.append(x[i])
    else:
      m.append(x[i])
  if len(m)==n:
    print("YES")
  else:
    print("NO")

 t=int(input())
 for i in range(t):
    k=int(input())
    sen=input()
    raf(k,sen)
 
 
 
```

```python
n=int(input())
x=sorted(map(int,input().split()))
q=int(input())
q_list=[]
for i in range(q):
  f=int(input())
  if f<x[0]:
    q_list.append(0)
  elif f>x[-1]:
    q_list.append(len(x))
  else:
    sum=0
    for j in x:
      if f>=j:
        sum=sum+1
      else:
        break
    q_list.append(sum)  
print(*q_list,sep='\n')
```

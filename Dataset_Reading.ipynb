`vitaldb.vital_recs(ipath, track_names=None, interval=None, return_timestamp=False, return_datetime=False, return_pandas=False, exclude=None)`

Parameter list

ipath: vital file path to read<br>
track_names: list of track names, eg) ['SNUADC/ECG', 'Solar 8000/HR'],<br>
interval: interval of each sample. if None, maximum resolution. if there is no wave tracks, 1 sec<br>
return_timestamp: return unix timestamp<br>
return_datetime: return datetime of each sample at first column<br>
return_pandas: return pandas dataframe<br>
Return value
numpy array of samples that have the same interval specified by the input argument.

```python
vitaldb.vital_recs(ipath, track_names='SNUADC/ART', return_pandas=True)
```

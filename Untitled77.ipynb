```python
import numpy as np
import pandas as pd

dataset = pd.read_csv("https://raw.githubusercontent.com/RafsanJany-44/Research-NREM-REM/main/dataset/O2/REM_NREM_O2.csv")
```

```python
dataset.head(10)
```

```python
dataset.shape
```

```python
dataset["Sleep_Stage"].unique()
```

#Split

```python
X = dataset.loc[:,dataset.columns != "Sleep_Stage"]

Y = dataset["Sleep_Stage"]
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2,random_state = 0)
```

#Training

### N Estimator Tunning

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

for i in range(1,201):
  clf = GradientBoostingClassifier(n_estimators = i)

  clf.fit(X_train,Y_train)

  y_pred = clf.predict(X_test)
  print("Estimator: ",i," Accuracy: ",accuracy_score(Y_test,y_pred))
```

### Depth Tunning

```python
for i in range(1,201):
  clf = GradientBoostingClassifier(max_depth = i)

  clf.fit(X_train,Y_train)

  y_pred = clf.predict(X_test)
  print("max_depth: ",i," Accuracy: ",accuracy_score(Y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,y_pred))
```

#Accuracy Matrics

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

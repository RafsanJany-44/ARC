#Marging `HR_Recovery_S01-S56` File

```python
no_need = ["patient_id","Condition","Segment"]



for i in list(hr_recovery.columns):
  if i not in no_need:
    hr_aee[i] = ''




for hr_aee_index in tqdm(range(hr_aee.shape[0])):

  p_id = r_aee_subject = hr_aee["patient_id"][hr_aee_index]
  condition = r_aee_subject = hr_aee["Condition"][hr_aee_index]

  for i in list(hr_recovery.columns):
    if i not in no_need:
      hr_aee[i][hr_aee_index] = np.nanmean(list(hr_recovery.query("patient_id == '"+p_id+"' and 	Condition == '"+condition+"'")[i]))
```

#Marging `Fitabase_combined_data_S01-S56` File

```python
no_need = ["patient_id","Condition","Activity","Epoch Duration(min)","Epoch"]
objet_type_col = ["fft_window","fft_interpolation","ar_interpolation"]


for i in list(py_hrv.columns):
  if i not in no_need:
    hr_aee[i+'_Extreme'] = ''
    hr_aee[i+'_Heavy'] = ''
    hr_aee[i+'_Moderate'] = ''
    hr_aee[i+'_Light'] = ''






for hr_aee_index in tqdm(range(hr_aee.shape[0])):
  p_id = r_aee_subject = hr_aee["patient_id"][hr_aee_index]
  condition = r_aee_subject = hr_aee["Condition"][hr_aee_index]

  for i in list(py_hrv.columns):
      if i not in no_need and i not in objet_type_col:
        hr_aee[i+'_Extreme'][hr_aee_index] = np.nanmean(list(py_hrv.query("patient_id == '"+p_id+"' and 	Condition == '"+condition+"' and 	Activity == 'Extreme'")[i]))
        hr_aee[i+'_Heavy'][hr_aee_index] = np.nanmean(list(py_hrv.query("patient_id == '"+p_id+"' and 	Condition == '"+condition+"' and 	Activity == 'Heavy'")[i]))
        hr_aee[i+'_Moderate'][hr_aee_index] = np.nanmean(list(py_hrv.query("patient_id == '"+p_id+"' and 	Condition == '"+condition+"' and 	Activity == 'Moderate'")[i]))
        hr_aee[i+'_Light'][hr_aee_index] = np.nanmean(list(py_hrv.query("patient_id == '"+p_id+"' and 	Condition == '"+condition+"' and 	Activity == 'Light'")[i]))





```

#Marging `Wearables-REDCap-CRF_S01-S56` File

```python
no_need=["patient_id", "record_id2"]

for i in list(all_info.columns):
  if i not in no_need:
    hr_aee[i] = ''




def all_info_back(p_id):
  li = []
  for i in list(all_info.columns):
    if i not in no_need:
      li.append((i,list(all_info.query("patient_id == '"+p_id+"'")[i])))
  return li




for hr_aee_index in tqdm(range(hr_aee.shape[0])):
  hr_aee_subject = hr_aee["patient_id"][hr_aee_index]
  infos = all_info_back(hr_aee_subject)

  for i in infos:
    hr_aee[i[0]][hr_aee_index] = i[1][0]
```
